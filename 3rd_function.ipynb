## Working function #3
### Here I created a dictionary with 2 keys (max_ma and min_ma)
#### These variables are used to describe the timeline of the species existence

```python
def makeProboscideaDict3(filename):
    # this function opens the data file 
    ProboscideaDict = {}


    in_file= open(filename, "r",encoding = "ISO-8859.15") # open the file
    all_records = in_file.readlines()[17:] # read the file (list)
    for line in all_records:
        mylist = line.strip('\n').replace('\"','') # removes unwanted quotation marks
        #mylist1 = mylist.replace('0','now')
        item = mylist.split(",") # split lines on comma (this produces another list)
        species = item[9]
        max_ma = item[14]
        min_ma = item[15]            
        ProboscideaDict[species] = (max_ma, min_ma)# stick species and occurence into ProboscideaDict
    return ProboscideaDict
    
```

```python
Proboscidea = makeProboscideaDict3("ideas/proboscidea.csv")
species = list(Proboscidea.keys())
for species in Proboscidea.keys():
#get the family in species
    mya = Proboscidea[species]# mya = million years ago
    print(species + " was known to exist during " + str(mya[0]) + " to "+str(mya[1])+ " million years ago")
```
